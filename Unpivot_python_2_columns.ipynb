```python
# Case 2 Dataframe
df1 = pd.DataFrame({'City': ['Houston', 'Austin', 'Hoover'],
                   'State': ['Texas', 'Texas', 'Alabama'],
                   'Name':['Aria', 'Penelope', 'Niko'],
                   'Mango':[4, 10, 90],
                   'Orange': [10, 8, 14], 
                   'Watermelon':[40, 99, 43],
                   'Gin':[16, 200, 34],
                   'Vodka':[20, 33, 18]},
                 columns=['City', 'State', 'Name', 'Mango', 'Orange', 'Watermelon', 'Gin', 'Vodka'])
df1
```

```python
dfx = df1.melt(id_vars=['City', 'State','Name'], 
              value_vars=['Mango', 'Orange', 'Watermelon'],
              var_name='Fruit', value_name='Pounds')
dfy = df1.melt(id_vars=['City', 'State','Name'], 
              value_vars=['Gin', 'Vodka'], 
              var_name='Drink', value_name='Ounces')
```

```python
dfx = dfx.set_index(['City', 'State','Name', dfx.groupby(['City', 'State','Name']).cumcount()])
dfy = dfy.set_index(['City', 'State', 'Name',dfy.groupby(['City', 'State','Name']).cumcount()])


```

```python
df3=(pd.concat([dfx, dfy],axis=1)
         .sort_index(level=2)
        .reset_index(level=3, drop=True)
        .reset_index())
```
